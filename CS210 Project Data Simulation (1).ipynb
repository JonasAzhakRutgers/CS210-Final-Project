```python
import numpy as np
import pandas as pd

np.random.seed(50)

n = 200 #this will represent the randomized factors for attorney's I'll randomize below

experience = np.random.randint(0, 31, size=n)
win_ratio = np.random.uniform(0.2, 0.9, size=n) #ratio strictly between 0.2-0.9
education_level = np.random.choice(['JSD', 'LLM', 'JD'], size=n, p=[0.1, 0.3, 0.6])

education_factored = np.array([1.0 if ed == 'JSD' else 0.85 if ed == 'LLM' else 0.7 for ed in education_level])
experience_factored = experience / 30

score = 0.5 * experience_factored + 0.3 * win_ratio + 0.2 * education_factored

case_result = np.where(score > 0.6, 1, 0)
print("Scores:", score)
print("Case Results:", case_result)
```

```python
dataframe = pd.DataFrame({
    'Experience_Years': experience,
    'Win_Loss_Ratio': win_ratio.round(2),
    'Education_Level': education_level,
    'Weighted_Score': score.round(2),
    'Case_Results': case_result
})
dataframe.head()
```

```python
dataframe.to_csv('attorney_case_dataset.csv', index=False)
```

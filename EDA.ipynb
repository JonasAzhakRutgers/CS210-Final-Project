```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('attorney_case_dataset.csv', na_values=[], keep_default_na=False)

print(df.head())
print(df.info())
print(df.describe())

print("\nMissing values:\n", df.isnull().sum())

# 1) Distribution of Case Results
sns.countplot(x='Case_Results', data=df)
plt.title("Distribution of Case Results")
plt.xlabel("Case Results")
plt.ylabel("Count")
plt.show()

# 2) Attorney Experience vs Case Results
sns.boxplot(x='Case_Results', y='Experience_Years', data=df)
plt.title("Attorney Experience vs Case Results")
plt.xlabel("Case Results")
plt.ylabel("Years of Experience")
plt.show()

# 3) Average Win/Loss Ratio by Education Level
sns.barplot(x='Education_Level', y='Win_Loss_Ratio', data=df)
plt.title("Average Win/Loss Ratio by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Avg. Win/Loss Ratio")
plt.xticks(rotation=45)
plt.show()

# 4) Weighted Score Distribution by Case Results
sns.violinplot(x='Case_Results', y='Weighted_Score', data=df)
plt.title("Weighted Score Distribution by Case Results")
plt.xlabel("Case Results")
plt.ylabel("Weighted Score")
plt.show()

# 5) Correlation Matrix of Numeric Features
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Numeric Features")
plt.show()
```
